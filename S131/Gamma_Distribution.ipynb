```python
from scipy.stats import gamma
import numpy as np
from matplotlib import pyplot as plt

a = 7
n = 1000 # number of events
bins = 30

events = gamma.rvs(a, size=1000)


plt.hist(events, bins = bins, range=[0, 20], density = True, label='events')
plt.xlabel('random variable')
plt.ylabel('probability')
plt.title('Events')
plt.legend()
plt.show()


#add noise
noise = np.random.normal(0, 1, size = [1000])
final_events = events + noise

plt.hist(final_events, bins = bins, range=[0, 20], density = True, label='Resulting hist')
plt.xlabel('random variable')
plt.ylabel('probability')
plt.title('Histogram with added noise')
plt.legend()
plt.show()

# Let's look at true distribution

x = np.linspace(gamma.ppf(0.01, a),
                gamma.ppf(0.99, a), 1000)

plt.plot(x, gamma.pdf(x, a), 'g', label='gamma pdf')
plt.xlabel('random variable')
plt.ylabel('probability')
plt.xlim([0, 20])
plt.legend()
plt.show()

train = gamma.fit(events, a)
pdf_gamma = gamma.pdf(x, train[0], train[1], train[2])

train_noisy = gamma.fit(final_events, a)
pdf_gamma_noisy = gamma.pdf(x, train_noisy[0], train_noisy[1], train_noisy[2])

plt.hist(final_events, bins = bins, range=[0, 20], density = True, label='Resulting hist')
plt.xlabel('random variable')
plt.ylabel('probability')
plt.plot(x, pdf_gamma, "g", linewidth = 4, label='True distribution') 
plt.plot(x, pdf_gamma_noisy, linewidth = 4, label='Fit Results') 
plt.legend()
```
